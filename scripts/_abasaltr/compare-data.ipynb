# Compare data files

```python
import pandas as pd
import numpy as np
from pandas_profiling import ProfileReport
```

```python
df_noise = pd.read_csv("../../data/raw/Sample_Noise_data.csv")
dfA = pd.read_csv("../../data/raw/Francis_etal_2012ProcB_pollination_data.csv")
dfB = pd.read_csv("../../data/raw/Francis_etal_2012ProcB_Seedling_data.csv")
dfC = pd.read_csv("../../data/raw/Francis_etal_2012ProcB_SeedRemoval_data.csv")
```

```python
#create specific columns from df - load
noise_columns = ["RECEIVED", "SRRECTYPE", "SR_TYPE_DESC", "WARD", "How_Recieved", "where_Recieved"]
df_noise_trans = df_noise[noise_columns].copy()

print(df_noise_trans.shape)
df_noise_trans.head()
```

```python
#create specific columns from df - load
dfA_columns = ["Site", "Pair_number", "Patch", "site_type", "amplitude_dB_A", "exposure"]
dfA_trans = dfA[dfA_columns].copy()

print(dfA_trans.shape)
dfA_trans.head()
```

```python
#create specific columns from df - load
dfB_columns = ["Plot_ID", "SITE", "site_type", "PSEEDLING", "CANOPY"]
dfB_trans = dfB[dfB_columns].copy()
dfB_trans.rename(columns= { 'SITE': 'Site' }, inplace=True)

print(dfB_trans.shape)
dfB_trans.head()
```

```python
#create specific columns from df - load
dfC_columns = ["Site", "Station", "amplitude_dB_A", "remaining", "mean_removed"]
dfC_trans = dfC[dfC_columns].copy()
dfC_trans.rename(columns= { 'amplitude_dB_A': 'amplitude_dB_C',
                          'Site': 'SiteC'}, inplace=True)
dfC_trans[['Station', 'Site']] = dfC_trans['Station'].str.split(' ', 1, expand=True)

print(dfC_trans.shape)
dfC_trans.head()
```

```python
dfAB = pd.merge(dfA_trans, dfB_trans)
print(dfAB.shape)
dfAB.head()
```

```python
dfABC = pd.merge(dfAB, dfC_trans)
print(dfABC.shape)
dfABC.head()
```

```python
profile_goal = ProfileReport(dfABC, title="Dryad Noise Pollution Overview", minimal=True)
profile_goal.to_file("dryad_noise_pollution_overview.html")
```
